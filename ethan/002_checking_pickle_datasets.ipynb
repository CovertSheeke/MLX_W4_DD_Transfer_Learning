```python
import pickle
from matplotlib import pyplot as plt

with open('/Users/ethanedwards/GitRepos/MLX_W4_DD_Transfer_Learning/flickr30k_processed_splits.pkl', 'rb') as f:
    dataset = pickle.load(f)
```

```python
import torch

# Get the test split from the dataset
test_data = dataset['train']
print(test_data)  # Print the first item to check its structure

# Plot the first 5 image-caption pairs
for i in range(5):
    image = test_data[i]['image']
    caption = test_data[i]['caption']
    
    # Convert embeddings back to tensors (HuggingFace datasets converts them to lists)
    img_embed = torch.tensor(test_data[i]['image_embed'])
    text_embed = torch.tensor(test_data[i]['text_embed'])
    
    print(f"Item {i+1}:")
    print(f"  Image embedding shape: {img_embed.shape}")
    print(f"  Text embedding shape: {text_embed.shape}")
    print(f"  Image embedding dims: {img_embed.shape[1]}")
    print(f"  Text embedding dims: {text_embed.shape[1]}")
    print(f"  Caption: '{caption[:50]}...'")
    print()
    
    plt.figure(figsize=(8, 6))
    plt.imshow(image)
    plt.title(f"Caption: {caption}")
    plt.axis('off')
    plt.show()
```
